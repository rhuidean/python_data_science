```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
import pandas as pd

import seaborn as sns
sns.set()

# create some data
rng =np.random.RandomState(0)
x = np.linspace(0,10,500)
y = np.cumsum(rng.randn(500,6),0)
```

```python
# plot 
plt.plot(x,y)
plt.legend('ABCDEF',ncol=2,loc='upper left')
```
