```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
fig = plt.figure() # create a plot figure
ax = plt.axes() # create axis

x = np.linspace(0,10,1000)
ax.plot(x,np.sin(x-0),color='blue')
ax.plot(x,np.sin(x-1),color='g')
ax.plot(x,np.sin(x-2),color='0.75')
ax.plot(x,np.sin(x-3),color='#FFDD44')
ax.plot(x,np.sin(x-4),linestyle='-.')
```
